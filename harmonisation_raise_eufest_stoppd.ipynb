```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 5)
pd.set_option('display.max_columns', None)

```

```python
##Directories.
#Folder name with original input data.
dir_input_eufest_baseline = 'EUFEST preproc/Baseline/Merged Datasets/'
dir_input_raise_baseline = 'RAISE preproc/Baseline/Merged Datasets/'
dir_input_stoppd_baseline = 'STOPPD preproc/Baseline/Merged Datasets/'
dir_input_eufest_labels = 'EUFEST preproc/Prediction set/Merged Datasets/'
dir_input_raise_labels = 'RAISE preproc/Prediction set/Merged Datasets/'
dir_input_stoppd_labels = 'STOPPD preproc/Prediction set/Merged Datasets/'
dir_output = "Harmonised All/"
##EUFEST

# NOTE THAT VIOLATORS HAVE BEEN REMOVED AND DATASETS WERE CREATED AS A LEFT MERGE TO V1, SO ALL HAVE THE SAME PATIENT IDs.

# Load clinical and sociodemographic data
df_clin_and_socio = pd.read_excel(dir_input_eufest_baseline + 'clinical_and_sociodemographic.xlsx')
df_blood_data = pd.read_excel(dir_input_eufest_baseline + 'lab_data.xlsx')
df_moa_mapping_data = pd.read_excel(dir_input_eufest_baseline + 'moa_mapping.xlsx')
df_cognitive = pd.read_excel(dir_input_eufest_baseline + 'cognitive.xlsx')
df_quality_of_life = pd.read_excel(dir_input_eufest_baseline + 'quality_of_life.xlsx')
df_prognosis = pd.read_excel(dir_input_eufest_baseline + 'prognosis.xlsx')
df_covariates = pd.read_excel(dir_input_eufest_baseline + 'covariates.xlsx')
df_labels = pd.read_excel(dir_input_eufest_labels + 'efficacy_labels.xlsx')

df_eufest = df_clin_and_socio.merge(df_blood_data, on='crfnr', how='left')
df_eufest = df_eufest.merge(df_moa_mapping_data, on='crfnr', how='left')
df_eufest = df_eufest.merge(df_cognitive, on='crfnr', how='left')
df_eufest = df_eufest.merge(df_quality_of_life, on='crfnr', how='left')
df_eufest = df_eufest.merge(df_prognosis, on='crfnr', how='left')
df_eufest = df_eufest.merge(df_covariates, on='crfnr', how='left')
df_eufest = df_eufest.merge(df_labels, on='crfnr', how='left')

##RAISE 
df_clin_and_socio = pd.read_excel(dir_input_raise_baseline + 'clinical_and_sociodemographic.xlsx')
df_health = pd.read_excel(dir_input_raise_baseline + 'health.xlsx')
df_cognitive = pd.read_excel(dir_input_raise_baseline + 'cognitive.xlsx')
df_quality_of_life = pd.read_excel(dir_input_raise_baseline + 'quality_of_life.xlsx')
df_covariates = pd.read_excel(dir_input_raise_baseline + 'covariates.xlsx')
df_labels = pd.read_excel(dir_input_raise_labels + 'efficacy_labels.xlsx')

df_raise = df_clin_and_socio.merge(df_health, on ='src_subject_id', how='left')
df_raise = df_raise.merge(df_cognitive, on='src_subject_id', how='left')
df_raise = df_raise.merge(df_quality_of_life, on='src_subject_id', how='left')
df_raise = df_raise.merge(df_covariates, on='src_subject_id', how='left')
df_raise = df_raise.merge(df_labels, on='src_subject_id', how='left')

##STOPPD
df_clin = pd.read_excel(dir_input_stoppd_baseline + 'clin.xlsx')
df_bio = pd.read_excel(dir_input_stoppd_baseline + 'bio.xlsx')
df_neurocog = pd.read_excel(dir_input_stoppd_baseline + 'neurocog.xlsx')
df_qol= pd.read_excel(dir_input_stoppd_baseline + 'qol.xlsx')
df_covariates = pd.read_excel(dir_input_stoppd_baseline + 'covariates.xlsx')
df_labels = pd.read_excel(dir_input_stoppd_labels + 'labels.xlsx')

df_stoppd = df_clin.merge(df_bio, on ='src_subject_id', how='left')
df_stoppd = df_stoppd.merge(df_neurocog, on ='src_subject_id', how='left')
df_stoppd = df_stoppd.merge(df_qol, on ='src_subject_id', how='left')
df_stoppd = df_stoppd.merge(df_covariates, on ='src_subject_id', how='left')
df_stoppd = df_stoppd.merge(df_labels, on ='src_subject_id', how='left')
```

```python
########################Calgary Depression Scale for Schizophrenia##############
##RAISE
column_mapping = {
    'calg1': 'V1CD01',
    'calg2': 'V1CD02',
    'calg3': 'V1CD03',
    'calg4': 'V1CD04',
    'calg5': 'V1CD05',
    'calg7': 'V1CD07',
    'calg8': 'V1CD08',
    'calg9': 'V1CD09',
    'calg_ts': 'V1CDTOTAL'
}

# Standardize the specified columns
for raise_col, eufest_col in column_mapping.items():

    #Standardise original EUFEST col.
    df_eufest[eufest_col] =  (df_eufest[eufest_col] - df_eufest[eufest_col].mean()) / df_eufest[eufest_col].std()

    #Change the name in RAISE col. 
    df_raise[eufest_col] = df_raise[raise_col].copy()
    
    # Standardize the RAISE cols.
    df_raise[eufest_col] = (df_raise[eufest_col] - df_raise[eufest_col].mean()) / df_raise[eufest_col].std()

##STOPPD
# Variable mapping.
column_mapping = {
    'hmdsd': 'V1CD01',
    #'calg2': 'V1CD02',
    #'calg3': 'V1CD03',
    #'hvwsf': 'V1CD04',
    'hvwsf': 'V1CD05',
    #'calg6': 'V1CD06',
    'hemin': 'V1CD07',
    'hsuic': 'V1CD08',
    #'hmdsd': 'V1CD09',
    'hamd_36': 'V1CDTOTAL'
}
# Iterate through the column_mapping and add new columns with the same values.
for stoppd_col, eufest_col in column_mapping.items():
    print(eufest_col)
    #Change name in STOP-PD col.
    df_stoppd[eufest_col] = df_stoppd[stoppd_col].copy()

    #Normalize columns
    df_stoppd[eufest_col] = (df_stoppd[eufest_col] - df_stoppd[eufest_col].mean()) / df_stoppd[eufest_col].std()
################################################################################

#################### Clinical Global Impression ################################
##RAISE
df_raise.rename(columns = {'cs16':'V1CGI_V1'}, inplace = True)

##STOP-PD
df_stoppd.rename(columns = {'cgi_si':'V1CGI_V1'}, inplace = True)

################################################################################


######################## Blood data ###########################################
##RAISE
# Name mapping
column_mapping = {
    'rsptc_no': 'V1l05',
    'rsphdl_no': 'V1l06',
    'rspldl_no': 'V1l07',
    'rsptrig_no': 'V1l09'
}

# Rename
df_raise.rename(columns=column_mapping, inplace=True)


##STOP-PD
# Name mapping
column_mapping = {
    'rsptc_no': 'V1l05',
    'rsphdl_no': 'V1l06',
    'rspldl_no': 'V1l07',
    'rsptrig_no': 'V1l09'
}

# Rename
df_stoppd.rename(columns=column_mapping, inplace=True)

###############################################################################


######################## Health ###############################################
#RAISE
column_mapping = {
    'vital_diabp': 'ph05',
    'vital_sysbp': 'ph04',
    'vital_pulse': 'ph06',
    'bmi': 'phbmi'
}

df_raise.rename(columns=column_mapping, inplace=True)

#Converting to imperial in eufest. 
df_eufest['phbmi'] =  (df_eufest['ph02']*2.20462) / ((df_eufest['ph01']/2.54)**2)*703


#STOPPD
column_mapping = {
    'vital_diabp': 'ph05',
    'vital_sysbp': 'ph04',
    'vital_pulse': 'ph06',
    'vt_bmi': 'phbmi'
}

df_stoppd.rename(columns=column_mapping, inplace=True)


###############################################################################


######################## DEMOGRAPHICS #########################################
##RAISE
column_mapping = {
    'interview_age': 'age',
    'sex': 'a003',
    'trt_set_1.0': 'a015_t_1',
    'trt_set_3.0': 'a015_t_2',
    'trt_set_2.0': 'a015_t_3',
    'trt_set_5.0': 'a015_t_4',
    'race_White': 'a022_1',
    'race_Black or African American': 'a022_2',
    'race_Asian': 'a022_3',
    'workyn': 'a030',
    'curres_1.0': 'a041'
}

#Change interview age from months to years. 
df_raise['interview_age'] = df_raise['interview_age']/12

# Renaming.
df_raise.rename(columns=column_mapping, inplace=True)

##STOPPD
column_mapping = {
    'interview_age': 'age',
    'sex': 'a003',
    'entry_status_1': 'a015_t_1',
    'entry_status_2': 'a015_t_2',
    'entry_status_3': 'a015_t_3',
    'entry_status_5': 'a015_t_4',
    'race_White': 'a022_1',
    #'race_Black or African American': 'a022_2',
    'race_Asian': 'a022_3',
    'race_American Indian/Alaska Native': 'a022_4',
    'eo4': 'a030',
    'live_1': 'a041' 

}

#Change interview age from months to years. 
df_stoppd['interview_age'] = df_stoppd['interview_age']/12

df_stoppd.rename(columns=column_mapping, inplace=True)

###############################################################################

####################Positive and Negative Syndrome Scale (PANSS)##############

#RAISE
# Variable mapping. 
column_mapping = {
    'pos_p1': 'V1pa01',
    'pos_p2': 'V1pa02',
    'pos_p3': 'V1pa03',
    'pos_p4': 'V1pa04',
    'pos_p5': 'V1pa05',
    'pos_p6': 'V1pa06',
    'pos_p7': 'V1pa07',
    'neg_n1': 'V1pa08',
    'neg_n2': 'V1pa09',
    'neg_n3': 'V1pa10',
    'neg_n4': 'V1pa11',
    'neg_n5': 'V1pa12',
    'neg_n6': 'V1pa13',
    'neg_n7': 'V1pa14',
    'gps_g1': 'V1pa15',
    'gps_g2': 'V1pa16',
    'gps_g3': 'V1pa17',
    'gps_g4': 'V1pa18',
    'gps_g5': 'V1pa19',
    'gps_g6': 'V1pa20',
    'gps_g7': 'V1pa21',
    'gps_g8': 'V1pa22',
    'gps_g9': 'V1pa23',
    'gps_g10': 'V1pa24',
    'gps_g11': 'V1pa25',
    'gps_g12': 'V1pa26',
    'gps_g13': 'V1pa27',
    'gps_g14': 'V1pa28',
    'gps_g15': 'V1pa29',
    'gps_g16': 'V1pa30',
    'panss_ptotal': 'V1ptotal',
    'panss_ntotal': 'V1ntotal',
    'panss_gtotal': 'V1gtotal',
    'panss_total': 'V1total'
}

# Standardize the specified columns
for raise_col, eufest_col in column_mapping.items():
    #Standardise original EUFEST col.
    df_eufest[eufest_col] =  (df_eufest[eufest_col] - df_eufest[eufest_col].mean()) / df_eufest[eufest_col].std()

    #Change the name in RAISE col. 
    df_raise[eufest_col] = df_raise[raise_col].copy()
    
    # Standardize the RAISE cols.
    df_raise[eufest_col] = (df_raise[eufest_col] - df_raise[eufest_col].mean()) / df_raise[eufest_col].std()

##STOPPD

#Take delusions from SAPS delusion scales. 
df_stoppd["saps_deltotal"] = df_stoppd.loc[:, 'saps_d1':'saps_d12'].sum(axis=1)

column_mapping = {
    'saps_deltotal': 'V1pa01',
    'bprs_conc': 'V1pa02',
    'bprs_hall': 'V1pa03',
    'bprs_exci': 'V1pa04',
    'bprs_gran': 'V1pa05',
    'bprs_susp': 'V1pa06',
    'bprs_host': 'V1pa07',
    'bprs_blun': 'V1pa08',
    'bprs_emot': 'V1pa09',
    #'neg_n3': 'V1pa10', #could be linked to bprs_emot
    #'neg_n4': 'V1pa11',
    #'neg_n5': 'V1pa12',
    #'neg_n6': 'V1pa13',
    #'neg_n7': 'V1pa14', #could be linked to bprs_conc
    'bprs_somc': 'V1pa15',
    'bprs_anxi': 'V1pa16',
    'bprs_guil': 'V1pa17',
    'bprs_tens': 'V1pa18',
    'bprs_mann': 'V1pa19',
    'bprs_depr': 'V1pa20',
    'bprs_motr': 'V1pa21',
    'bprs_unco': 'V1pa22',
    'bprs_unus': 'V1pa23',
    'bprs_diso': 'V1pa24',
    #'gps_g11': 'V1pa25',
    'hinsg': 'V1pa26', #Lack of insight in Hamilton
    #'gps_g13': 'V1pa27',
    #'gps_g14': 'V1pa28',
    #'gps_g15': 'V1pa29',
    #'gps_g16': 'V1pa30',
    #'panss_ptotal': 'V1ptotal',
    #'panss_ntotal': 'V1ntotal',
    #'panss_gtotal': 'V1gtotal',
    #'panss_total': 'V1total'
}

# Iterate through the column_mapping and add new columns with the same values.
for stoppd_col, eufest_col in column_mapping.items():
    #Change name in STOP-PD col.
    df_stoppd[eufest_col] = df_stoppd[stoppd_col].copy()

    #Normalize columns
    df_stoppd[eufest_col] = (df_stoppd[eufest_col] - df_stoppd[eufest_col].mean()) / df_stoppd[eufest_col].std()

################################################################################

########################SUBSTANCE USE###########################################
#RAISE
#Alcohol current use.
df_eufest['SA_Alcohol'] = df_eufest[['V1SAL06']].apply(lambda row: 1 if 1 in row.values else 0, axis=1)
df_raise['SA_Alcohol'] = df_raise['subus25']

#Cannabis.
df_eufest['SA_Cannabis'] = df_eufest[['V1SAL03']].apply(lambda row: 1 if 1 in row.values else 0, axis=1)
df_raise['SA_Cannabis'] = df_raise['ca824']

#Cannabis prev use.
df_eufest['SA_Cannabis_prev'] = df_eufest[['V1SA06']].apply(lambda row: 1 if 1 in row.values else 0, axis=1)
df_raise['SA_Cannabis_prev'] = df_raise['subus29']

#Other illicit current drug use.
df_eufest['SA_Other'] = df_eufest[['V1SAL08', 'V1SAL11', 'V1SAL14', 'V1SAL17', 'V1SAL20', 'V1SAL23']].apply(lambda row: 1 if 1 in row.values else 0, axis=1)
df_raise['SA_Other'] = df_raise['subus30']


#STOPPD
#In DSM, 1 is absent and 3 is present. 
df_stoppd['SA_Alcohol'] = df_stoppd['q068_al_mon'].apply(lambda row: 1 if row > 1 else (0 if row == 1 else np.nan))
df_stoppd['SA_Cannabis'] = df_stoppd['q072_can_mon'].apply(lambda row: 1 if row > 1 else (0 if row == 1 else np.nan))
df_stoppd['SA_Cannabis_prev'] = df_stoppd['q071_can_life'].apply(lambda row: 1 if row > 1 else (0 if row == 1 else np.nan))
df_stoppd['SA_Other'] = df_stoppd[['q070_sha_mon', 'q074_stim_mon', 'q076_op_mon', 'q078_coc_mon', 'q080_hal_mon', 'q082_poly_mon', 'q084_othsub_mon']].apply(lambda row: 1 if (row == 3).any() else 0, axis=1)

################################################################################


######################## QUALITY OF LIFE ######################################

#RAISE
eufest_vars = ['V1MA02', 'V1MA06', 'V1MA12', 'V1MA13', 'V1MA14', 'V1MA15', 'V1MA16']
raise_vars = ['qol13', 'qol03', 'qol05', 'qol02', 'qol09', 'sf01', 'srf019']
#Standardise by the max possible value and min possible value of each of the scores.  

#MANSA scores go from 1 to 7.
df_eufest[eufest_vars] = (df_eufest[eufest_vars]- 1)/(7-1)

#QOL scores go from 0 to 6. 
df_raise[['qol13', 'qol03', 'qol05', 'qol02', 'qol09']] = (df_raise[['qol13', 'qol03', 'qol05', 'qol02', 'qol09']] - 0)/(6-0)

#sf-12 scores go from 1 to 5, 1 meaning good and 5 meaning poor. 
df_raise['sf01'] = (df_raise['sf01'] - 1)/(5-1)

#Invert to consider the fact that the scale is inverted vs. MANSA. 
df_raise['sf01'] = 1 - df_raise['sf01']

#SRF scores go from 1 to 7. 
df_raise['srf019'] = (df_raise['srf019'] - 1)/(7-1)

#Get social satisfaction measure. 
df_eufest["social_satisf"] = df_eufest[['V1MA02', 'V1MA06', 'V1MA12', 'V1MA13', 'V1MA14']].sum(axis=1)
df_raise["social_satisf"] = df_raise[['qol13', 'qol03', 'qol05', 'qol02', 'qol09']].sum(axis=1)

df_eufest["general_health_satif"] = df_eufest["V1MA15"].copy()
df_raise["general_health_satif"] = df_raise["sf01"].copy()

df_eufest["mental_health_satif"] = df_eufest["V1MA16"].copy()
df_raise["mental_health_satif"] = df_raise["srf019"].copy()

#STOP-PD
df_stoppd["social_satisf"] = df_stoppd["socialact"].copy()
df_stoppd["general_health_satif"] = df_stoppd["mospcs"].copy()
df_stoppd["mental_health_satif"] = df_stoppd["mosmcs"].copy()

#Z-scores to be used across datasets. 
for var in ["social_satisf", "general_health_satif", "mental_health_satif"]:
    df_eufest[var] = (df_eufest[var] - df_eufest[var].mean()) / df_eufest[var].std()
    df_raise[var] = (df_raise[var] - df_raise[var].mean()) / df_raise[var].std()
    df_stoppd[var] = (df_stoppd[var] - df_stoppd[var].mean()) / df_stoppd[var].std()

###############################################################################


########################Neurocognitive perf####################################

#hand coordination total score. 
df_eufest['V1NEUHAND'] = df_eufest['V1NEU10']+df_eufest['V1NEU11']+df_eufest['V1NEU12']

#RAISE and STOPPD
#Normalise executive functioning variables. 
df_eufest['exec_functioning'] = (df_eufest['V1NEU04MINUS01'] - df_eufest['V1NEU04MINUS01'].mean()) / df_eufest['V1NEU04MINUS01'].std()
df_raise['exec_functioning'] = (df_raise['bacs_tl_total'] - df_raise['bacs_tl_total'].mean()) / df_raise['bacs_tl_total'].std()
df_stoppd['exec_functioning'] = (df_stoppd['stroop_interference_score'] - df_stoppd['stroop_interference_score'].mean()) / df_stoppd['stroop_interference_score'].std()

#Processing speed variables. 
df_eufest['processing_speed'] = (df_eufest['V1NEUHAND'] - df_eufest['V1NEUHAND'].mean()) / df_eufest['V1NEUHAND'].std()
df_raise['processing_speed'] = (df_raise['bactmts'] - df_raise['bactmts'].mean()) / df_raise['bactmts'].std()
df_stoppd['processing_speed'] = (df_stoppd['stroop1'] - df_stoppd['stroop1'].mean()) / df_stoppd['stroop1'].std()

###############################################################################


############################# LABELS ####################################

#RAISE
column_mapping = {
    'MJust6CD_PSD_CLASS': 'V367CD_PSD_CLASS',
    'MJust6CD_PSD_REG': 'V367CD_PSD_REG'
}

df_raise.rename(columns=column_mapping, inplace=True)

#STOP-PD
column_mapping = {
    'DEP_HAMD_REMISSION_12W': 'V367CD_PSD_CLASS',
    'DEP_TOTAL_12W': "V367CD_PSD_REG"
}

df_stoppd.rename(columns=column_mapping, inplace=True)

###############################################################################

############################# Patient ID ######################################
df_raise.rename(columns={'src_subject_id':'crfnr'}, inplace=True)
df_stoppd.rename(columns={'src_subject_id':'crfnr'}, inplace=True)
###############################################################################

#Variables that were harmonised. 
variables = ['crfnr', 'V1CD01', 'V1CD05', 'V1CD07', 'V1CD08', 'V1CDTOTAL', 'V1CGI_V1', 'V1l05', 'V1l06', 'V1l07', 'V1l09',  
             'ph05', 'ph04', 'ph06', 'phbmi', 'age', 'a003', 'a015_t_1', 'a015_t_2', 'a015_t_3', 'a015_t_4', 'a022_1', 
             'a022_3', 'a030', 'a041', "V1pa01", "V1pa02", "V1pa03", "V1pa04", "V1pa05", "V1pa06", "V1pa07", "V1pa08", "V1pa09",  
             "V1pa15", "V1pa16", "V1pa17", "V1pa18", "V1pa19", "V1pa20", "V1pa21", "V1pa22", "V1pa23", "V1pa24", "V1pa26",  
             'SA_Alcohol', 'SA_Cannabis', 'SA_Cannabis_prev', 'SA_Other', "social_satisf", "general_health_satif",  
             "mental_health_satif", 'exec_functioning', 'processing_speed']




labels = ['V367CD_PSD_CLASS', 'V367CD_PSD_REG']

vars_to_add = variables + labels

#Save to excel.
df_eufest = df_eufest[[col for col in vars_to_add if col in df_eufest.columns]].copy()
df_raise = df_raise[[col for col in vars_to_add if col in df_raise.columns]].copy()
df_stoppd = df_stoppd[[col for col in vars_to_add if col in df_stoppd.columns]].copy()

#Add name of sample to ID
df_eufest["crfnr"] = "EUFEST" + df_eufest["crfnr"].astype(str)
df_raise["crfnr"] = "RAISE" + df_raise["crfnr"].astype(str)
df_stoppd["crfnr"] = "STOPPD" + df_stoppd["crfnr"].astype(str)

#Save EUFEST and RAISE.
df_eufest.to_excel(dir_output + 'harmonised_eufest_data.xlsx', index = None)
df_raise.to_excel(dir_output + 'harmonised_raise_data.xlsx', index = None)
df_merged = pd.concat([df_eufest, df_raise])
df_merged.to_excel(dir_output + 'harmonised_eufestraise_data.xlsx', index = None)

#Save STOP-PD
df_stoppd.to_excel(dir_output + "harmonised_stoppd_data.xlsx", index = None)
```
